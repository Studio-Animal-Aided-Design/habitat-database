```python
import pandas as pd

file_path = '../Grafiken_Zielarten_Lebenszyklen.xlsx'
xls = pd.ExcelFile(file_path)

# Checking sheet names to understand the file's structure
sheet_names = xls.sheet_names
sheet_names
```

```python
images_df = pd.read_excel(xls, sheet_name='Tabelle1')
images_df.head()
```

```python
# Clean the column names for snake_case and English translation
images_df.columns = [
    'common_name', 'scientific_name', 'portrait_image', 'lifecycle_image'
]

images_df.head()



```

```python
# transform into speciesImage records

# one for portrait_image and one for lifecycle_image

# Create the new DataFrame
portrait_df = images_df[['scientific_name', 'portrait_image']].copy()
portrait_df.columns = ['species', 'image_url']
portrait_df['image_type'] = 'portrait'

lifecycle_df = images_df[['scientific_name', 'lifecycle_image']].copy()
lifecycle_df.columns = ['species', 'image_url']
lifecycle_df['image_type'] = 'lifecycle'

# Concatenate both DataFrames
new_df = pd.concat([portrait_df, lifecycle_df], ignore_index=True)

# Hinzufügen einer 'id'-Spalte mit fortlaufenden Ganzzahlen
new_df.insert(0, 'id', range(1, len(new_df) + 1))
new_df.insert(3, 'image_alt', None)
new_df.insert(3, 'attribution', None)

new_df.head()
```

```python
# Exportieren der bereinigten Daten als CSV-Datei
output_path = './out/species-images.csv'
new_df.to_csv(output_path, index=False)

print(f'Die Datei wurde erfolgreich exportiert nach: {output_path}')
```
